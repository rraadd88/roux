```python
import matplotlib.pyplot as plt
```

### Step plot

**Demo data**

```python
import pandas as pd
data=pd.DataFrame({
    'sample size':[100,30,60,50,30,20,25],
    })
data['step name']='step#'+(data.index+1).astype(str)
data
```

#### Plot

```python
from roux.viz.line import plot_steps
ax=plot_steps(
    data,
    col_step_name='step name',
    col_step_size='sample size',
    )
```

## bezier lines

```python
from roux.viz.line import plot_bezier
# %run ../../roux/roux/viz/line.py
fig,ax=plt.subplots(figsize=[2,2])
for direction in ['h','v']:
    for off_guide in [0.25,0.75,1]:
        plot_bezier(
            [0,0],
            [1,1],
            direction=direction,
            off_guide=off_guide,
            test=True,
        )
ax.axis('off')
_=plt.legend(bbox_to_anchor=[1,1],title='guides')
```

```python
## required functions
from roux.viz.io import read_plot
```

```python
## parameters
### input path
input_path='outputs/roux_viz_io//plot_saved.png'
```

## Base theme

```python
_=read_plot(input_path)
```

## Theming

```python
from roux.viz.theme import set_theme #noqa
```

```python
ax=read_plot(input_path)

from roux.viz.ax_ import format_ax
_=format_ax(
    ax=ax,
    rotate_ylabel=True,
)
_=ax.legend(bbox_to_anchor=[1.5,1])
```

## Formatting

```python
ax=read_plot(input_path)

ax.set(
    xlabel='xlabel is longer than usual',
    ylabel='ylabel is longer than usual'
)
from roux.viz.ax_ import format_ax
_=format_ax(
    ax=ax,
    textwrap_width=20,
)
_=ax.legend(bbox_to_anchor=[1.5,1])
```

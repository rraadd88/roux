## 📈 Distribution plots.

```python
# installing the required roux subpackage
# !pip install roux[viz]
# loading requirements
import seaborn as sns
import matplotlib.pyplot as plt
# ## for saving plots
# from roux.viz.io import to_plot
```

### Demo data

```python
df1=sns.load_dataset('iris')
# df1=df1.loc[df1['species'].isin(['setosa','virginica']),:]
df1['id']=range(len(df1))
df1.head(1)
```

### Paired horizontal (preferred)

```python
fig,ax=plt.subplots(figsize=[2,2])
from roux.viz.dist import plot_dists
ax=plot_dists(df1,x='sepal_length',y='species',colindex='id',kind=['box','strip'],order=['setosa','virginica'])
# to_plot('./outputs/roux_viz_dists/plot_dists_h.png')
```

### Paired vertical

```python
fig,ax=plt.subplots(figsize=[2,2])
from roux.viz.dist import plot_dists
ax=plot_dists(
    df1,
    x='species',
    y='sepal_length',
    colindex='id',
    kind=['box','strip'],
    order=['setosa','virginica'],
    offs_pval=dict(y=0.25),
    )
sns.despine()
# to_plot('./outputs/roux_viz_dists/plot_dists_v.png')
```

### >2 categories

```python
fig,ax=plt.subplots(figsize=[3,2])# larger figure size
from roux.viz.dist import plot_dists
ax=plot_dists(
    df1,
    x='sepal_length',
    y='species',
    colindex='id',
    kind=['box','strip'],
    offs_pval=dict(y=-0.25),
    showmeans=False,
    )
# plt.tight_layout()
# ax.set(
#     xlim=(ax.get_xlim()[0],ax.get_xlim()[1]+1)
# )
# to_plot('./outputs/roux_viz_dists/plot_dists_v_gt2.png')
```

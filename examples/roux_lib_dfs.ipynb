## ⌗⌗ Paired Tables/Dataframes.

## Demo data

```python
nrows=5
```

```python
## left
import numpy as np
np.random.RandomState(1)
import pandas as pd
data=pd._testing.makeCustomDataframe(
    nrows=nrows,
    ncols=2,
).reset_index(drop=True)
data.columns=['a 1','a 2']
data=data.assign(left='left')
data
```

```python
## right
data2=pd.concat(
    [data.loc[:,['a 1']].sample(nrows-1).rename(columns={'a 1':'a'}),
    data.loc[:,['a 2']].sample(nrows-1).rename(columns={'a 2':'a'})],
    axis=0).assign(
    **{
        'right':'right',
        'right common':lambda df: df['a'].str[:2],
      }
    )
# data2['right common']=range(len(data2))
data2
```

## Import `r`oux-`d`ataframe attributes

```python
import roux.lib.dfs as rd
```

## Mergeing

```python
out=data.rd.merge_paired(
    data2,
    left_ons=['a 1','a 2'],
    right_on='a',
    how='inner',
    validates=['1:1','1:1']    
)
out
```

```python
# for testing
assert out.shape==(3, 7), out.shape
```

```python
out=data.rd.merge_paired(
    data2,
    left_ons=['a 1','a 2'],
    right_on='a',
    how='left',
    validates=['m:1','m:1']
)
out
```

```python
# for testing
assert out.shape==(5, 7), out.shape
```

```python
out=data.rd.merge_paired(
    data2,
    left_ons=['a 1','a 2'],
    right_on='a',
    right_ons_common=['right common'],
    how='inner',
    validates=['m:1','m:1']
)
out
```

```python
# for testing
assert out.shape==(3, 6), out.shape
```

### Demo data and parameters

```python
# demo data
import seaborn as sns
data=sns.load_dataset('iris')
```

```python
## parameters
### output path
output_path='tests/output/plot/plot_modified.png'
### for carrying out modifications in the plot, as shown in the next section.
kws_plot_modified=dict(y='petal_length')
```

### Saving plots with the source data

```python
# import helper functions
from pathlib import Path 
import matplotlib.pyplot as plt
from roux.viz.io import begin_plot,to_plot,read_plot

## parameters for the plot
kws_plot=dict(
    x='sepal_length',
    y='sepal_width',
)
## path to the saved plot
plot_path=output_path.replace('modified','saved')
```

```python
## begin the plot. (In the background, it initiates the logging of the code from this cell.)
begin_plot()
fig,ax=plt.subplots(figsize=[3,3])
sns.scatterplot(
    data=data,
    hue='species',
    ax=ax,
    **kws_plot,
    )
## save the plot
to_plot(
    plot_path,# path to the saved plot
    data=data, # source data
    kws_plot=kws_plot,# plotting parameters
    validate=True, #validates reusability and optimizes the performance for the next use
    )
assert Path(plot_path).exists()
```

### "Reading" the plots

```python
ax=read_plot(plot_path)
```

### "Reading" the plot and modifying its parameters

```python
## the parameters for carrying out the modifications in the plot
kws_plot_modified
```

```python
read_plot(plot_path,kws_plot=kws_plot_modified,title='modified')
plt.savefig(output_path)
```

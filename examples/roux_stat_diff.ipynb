**Demo data**

```python
import numpy as np
import pandas as pd #noqa
# np.random.seed(0)
import seaborn as sns
df1=(
    sns.load_dataset('iris')
    .assign(
        id=lambda df: range(len(df)),
        group1=lambda df: np.random.choice([0,1],len(df)),
        group2=lambda df: np.random.choice([0,1],len(df)),
    )
)
df1.head()
```

## Differences

### Single values, no groups

```python
from roux.stat.diff import get_stat
get_stat(
    df1,
    colsubset='species',
    colvalue='sepal_length',
    colindex='id',
)
```

### Multiple values, no groups

```python
from roux.stat.diff import get_stats
get_stats(
    df1,
    colsubset='species',
    cols_value=['sepal_length','sepal_width'],
    colindex='id',    
)
```

### Mutiple values, with groups

```python
from roux.stat.diff import get_stats_groupby
get_stats_groupby(
    df1,
    colsubset='species',
    cols_value=['sepal_length','sepal_width'],
    colindex='id',
    cols_group=['group1','group2']
).head(1)
```
